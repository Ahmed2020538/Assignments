```python
import sqlite3 as S
con = S.connect("Books_data_sets.db")
cur = con.cursor()

def Extract_data(sql , data):
    global con , cur
    cur.execute(sql , data)
    rows = cur.fetchall()
    for B_Id , B_Name , B_Price , B_Rate ,  C_Id in rows :
        print(f"The B_Id :: {B_Id}\nThe B_Title :: {B_Name}\nThe B_Price :: {B_Price} £\nThe B_Rate :: {B_Rate}\nC_Id :: {C_Id}\n\n-----------------------------")
    con.commit()
```

 1:: Write a sql query to get books that has rate >=3 and has 'Mr' in its name.

```python
Extract_data("select * from Books where B_Rate >= ? and B_Name like ?" ,[3 , "%Mr%"])
```

2 :: Write a sql query to get books that has rate >=3 or price > 20.**

```python
Extract_data("select * from Books where B_Rate >= ? or B_Price > ?" ,[3 , 20])
```

3:: Write a sql query to get books that has rate not 3.**

```python
Extract_data("select * from Books where B_Rate != ?" ,[3])
```

4:: Write a sql query to get books that has price between 10 and 40 and has rate 3 or 4 or 1.**

```python
Extract_data("select * from Books where B_Price between ? and ? and B_Rate in(?,?,?) " ,[10,40,3,4,1])
```

5:: Write a sql query to get the top 5 most expensive books (order books desc with price column and then limit the result to the first 5).**

```python
Extract_data("select * from Books order by B_Price desc limit ?" , [5])
```

6:: Write a sql query to get the 3rd 10 books in the books table.**

```python
Extract_data("select * from Books  limit ? offset ?" , [10,3])
```
