1) Using titanic_full.csv dataset, extract Mr,Miss,Mrs,etc.. from 'Name' column and use it to create new column and call it 'Title', then plot a countplot showing the count for each title in the dataset.

```python
df = pd.read_csv("titanic_full.csv")
df
```

```python
def Title_Name (name) :
    if "Mr." in name :
        return "MR" 
    elif "Mrs." in name :
        return "MRS"
    elif "Miss." in name :
        return "MISS"
    else : 
        return "OTHER"
```

```python
df["Title"] = df["Name"].apply(Title_Name)
```

```python
df[["Title"]]
```

```python
sns.countplot( x = "Title" , data = df )
```

1) Using visitor-interests.csv dataset, from the first 50 row extract Browser, OS, Device data from 'UserAgent' column and use it to create new columns for Browser, OS, Device.

```python
df = pd.read_csv("visitor-interests.csv")
df 
```

```python
df = df.head(50)
df
```

```python
def User_Agent_data(ua) :
    res = UA.parse(ua)
    Browser_Family  = res.browser.family
    Browser_Version = res.browser.version
    OS_Family       = res.os.family
    OS_Version      = res.os.version
    device_Family   = res.device.family
    device_Model    = res.device.family
    device_Brand    = res.device.brand

    return pd.Series((Browser_Family , Browser_Version , OS_Family , OS_Version , device_Family , device_Model , device_Brand ))
```

```python
df[["Browser_Family" , "Browser_Version" , "OS_Family" , "OS_Version" , "device_Family" , "device_Model" 
    , "device_Brand" ]] = df["UserAgent"].apply(User_Agent_data)
```

```python
df[["Browser_Family" , "Browser_Version" , "OS_Family" , "OS_Version" , "device_Family" , "device_Model" 
    , "device_Brand" ]] 
```

```python

fig =sns.countplot(x = "Browser_Family" , data = df)
fig.set_xticklabels( labels =df["Browser_Family"].value_counts().index  , rotation = 90)
```

```python
sns.countplot(x = "OS_Family" , data = df)
```

```python
Top_User_device_Models = df["device_Model"].value_counts()[:6]
Top_User_device_Models



```

```python
fig =sns.countplot(x = "device_Model" , data = df[df["device_Model"].isin(Top_User_device_Models.index)] )
fig.set_xticklabels( labels = Top_User_device_Models.index  , rotation = 90)
```

2) Using visitor-interests.csv dataset, from the first 50 row extract latitude, longitude from 'IP' column.

```python
df
```

```python
def IP_User_data(ip) :
    try :
        res = IP.get( ip , api_key = "free")
        City      = res.city
        Country   = res.country
        
        return pd.Series(( City , Country))
    except :
        return "Missing"
```

```python
df[["City" , "Country"]] = df["IP"].apply(IP_User_data)
```

```python
df[[ "City" , "Country"]]
```

```python
sns.countplot("City" , data = df , palette = "Dark2")
plt.xticks(rotation = 90)
```

```python
sns.countplot("Country" , data = df , palette = "Dark2")
```
