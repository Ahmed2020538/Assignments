**Use <code>for</code>, .split(), and <code>if</code> to create a Statement that will print out words that start with 's':**

```python
st = 'Print only the words that start with s in this sentence'
s = st.split(" ")
list(s)
for i in s :
    if i.startswith("s") : 
        print("=====================================")
        print(f"The Word is start with s is :: {i}")
        print("=====================================")
        print("The list of letters to The Word is ::" ,list(i))
    
```

**Use range() to print all the even numbers from 0 to 10.**

```python
l = list(range(11))
for i in l :
    if i % 2 == 0 :
        print(f"That Num :: {i} :: is Even  ")
        print("===============================")
    else : 
        print(f"That Num :: {i} :: is Odd  ")
        print("==============================")
        
        
```

**Use a List Comprehension to create a list of all numbers between 1 and 50 that are divisible by 3.**

```python
l = list(range(1,51))
for i in l :
    if i % 3 == 0 :
        print(f"That Num :: {i} :: is divisible by 3")
        print("=======================================")
    else :
        print(f"That Num :: {i} :: is not divisible by 3")
        print("=========================================")
        
```

**Go through the string below and if the length of a word is even print "even!"**

```python
st = 'Print every word in this sentence that has an even number of letters'
s = st.split(" ")
list(s)
for i in s :
    if len(i) % 2 == 0 :
        print(f"The length of Word :: {i} :: is {len(i)}")
        print(f"The Length of that Word  :: {i}  :: is Even   ^_^ ")
        print("=============================================")
    else :
        print(f"The length of Word :: {i} :: is {len(i)}")
        print(f"The Length of that Word  :: {i}  :: is Odd  ^_^")
    print("================================================")
```

Write a program that prints the integers from 1 to 100. But for multiples of three print "Fizz" instead of the number, and for the multiples of five print "Buzz". For numbers which are multiples of both three and five print "FizzBuzz".

```python
l = list(range(1,101))
for i in l :
    if i % 3 == 0 and i % 5 == 0 :
        print("Fizz BuZZ" + " ^_^")
        print("==================")
    elif i % 3 == 0 :
        print("FiZZ" + " ^_^")
        print("==================")
    elif i % 5 == 0 :
        print("BuZZ" , " ^_^")
        print("==================")
    else :
        print("GaMe OveR Man" + " ^_^")
        print("==================")
```

**Use List Comprehension to create a list of the first letters of every word in the string below:**

```python
st = 'Create a list of the first letters of every word in this string'
s = st.split(" ")
list(s)
for i in s :
    print(f"The Word is :: {i}")
    print(f"The first letter of :: {i} :: is :: {i[0]}")
    print("===========================================")
```
